```python
df = con.execute("""
SELECT *
FROM daily_activity
WHERE cage_id = 4917
ORDER BY animal_id, date
""").df()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for animal_id, g in df.groupby("animal_id"):
    plt.plot(g["date"], g["active_seconds"], alpha=0.7)

plt.title("Cage 4917: Daily Activity per Animal")
plt.xlabel("Date")
plt.ylabel("Active seconds")
plt.show()
```

```python
df_norm = df.copy()
df_norm["z_activity"] = (
    df_norm.groupby("animal_id")["active_seconds"]
    .transform(lambda x: (x - x.mean()) / x.std())
)

plt.figure(figsize=(12,6))
for animal_id, g in df_norm.groupby("animal_id"):
    plt.plot(g["date"], g["z_activity"], alpha=0.7)

plt.title("Cage 4917: Normalized Activity (Z-score)")
plt.ylabel("Z-scored activity")
plt.show()
```

```python
import seaborn as sns

pivot = df_norm.pivot(
    index="animal_id",
    columns="date",
    values="z_activity"
)

plt.figure(figsize=(14,6))
sns.heatmap(pivot, cmap="coolwarm", center=0)
plt.title("Cage 4917: Activity Alignment Heatmap")
plt.xlabel("Date")
plt.ylabel("Animal ID")
plt.show()
```
